VISUALIZATION TECHNIQUES (E.G., BOXPLOTS)

STEP 36: Identify outliers using IQR 

```python
# Using IQR Method 
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns 
outliers_iqr = {} 
for col in numeric_cols: 
    Q1 = df[col].quantile(0.25) 
    Q3 = df[col].quantile(0.75) 
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR 
    outliers_iqr[col] = df[(df[col] < lower_bound) | (df[col] > 
    upper_bound)][col] 
    print(f"\nOutliers in {col} using IQR:") 
    print(outliers_iqr[col].head()) 
```

STEP 37: Identify outliers using Z-score 


```python
from scipy import stats 
outliers_zscore = {} 
for col in numeric_cols: 
    # Drop NaNs and compute Z-scores 
    col_data = df[col].dropna() 
    z_scores = stats.zscore(col_data) 
    # Create a boolean mask aligned with the dropped series 
    mask = abs(z_scores) > 3 
    # Use the index of the filtered col_data to get values from original df 
    outlier_indices = col_data[mask].index 
    outliers_zscore[col] = df.loc[outlier_indices, col] 
    print(f"\nOutliers in {col} using Z-score:") 
    print(outliers_zscore[col].head()) 
```

STEP 38: Using boxplots for outlier detection 

```python
df_copy = df.copy() 
plt.figure(figsize=(15, 30)) 
for i, col in enumerate(numeric_cols, 1): 
    plt.subplot(5, 4, i) 
    sns.boxplot(y=df_copy[col]) 
    plt.title(col) 
    plt.tight_layout() 
    plt.show()
```

STEP 39: Capping for handling extreme values 

```python
for col in numeric_cols: 
    lower_cap = df_copy[col].quantile(0.1)  # 5th percentile 
    upper_cap = df_copy[col].quantile(0.9)  # 95th percentile 
    df_copy[col] = df_copy[col].clip(lower=lower_cap, upper=upper_cap) 
    print(f"\nCapped {col} between {lower_cap} and {upper_cap}")
```

STEP 40: Removing extreme values 


```python
for col in numeric_cols: 
    Q1 = df_copy[col].quantile(0.25) 
    Q3 = df_copy[col].quantile(0.75) 
    IQR = Q3 - Q1 
    lower_bound = Q1 - 1.5 * IQR 
    upper_bound = Q3 + 1.5 * IQR 
    df_copy = df_copy[(df_copy[col] >= lower_bound) & (df_copy[col] <= upper_bound)] 
    print(f"\nRemoved outliers in {col} outside {lower_bound} and {upper_bound}") 
```

STEP 41: Log Transformation for extreme values

```python
for col in numeric_cols: 
    df_copy[f'{col}_log'] = np.log1p(df_copy[col])  # log1p to handle zeros 
    print(f"\nApplied log transformation to {col}") 
```
